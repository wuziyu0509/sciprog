# Basel问题简介

   巴塞尔问题是一个著名的级数问题，问题首先由皮耶特罗·门戈利在1644年提出，这个问题是就精确计算所有平方数的倒数的和，也就是以下级数的和：
   $\sum\limits_{n=1}^{\infty}{1 \over n^2} $ \
   最后由欧拉于1735年解决，证明其收敛于${\pi^2 \over 6}$

# 数值验证 Basel问题

```python
import matplotlib.pyplot as plt
import math
plt.figure(num=1,figsize=(20,8),dpi=100) #铺设画布
sum = 0
MAXN = 100
yline = math.pi*math.pi/6
plt.axhline(yline,0,MAXN+1) #画y = pi^2/6
for n in range (1,MAXN):
    sum = sum + 1/(n*n)
    plt.scatter(n,sum) # 画出散点图
plt.show()
```

可以发现，贝塞尔函数确实收敛于${\pi^2 \over 6}$

## 讨论误差和 n 的关系

```python
plt.figure(num=2,figsize=(20,8),dpi=100) #铺设画布
sum = 0
for n in range (1,MAXN):
    sum = sum + 1/(n*n)
    plt.scatter(n,yline-sum) # 画出误差图
plt.show()
```

从图中我们可以发现当n越大，此级数的和越接近${\pi^2 \over 6}$
